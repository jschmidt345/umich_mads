# Array Operations

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b
print(c)

d = a * b
print(d)

# elementwise
```

```python
farenheit = np.array([0, -10, -5, -15, 0])

# formula for conversion is ((F - 32)) * 5 / 9
celcius = (farenheit - 32) * (5 / 9)
celcius
```

```python
celcius > -20
```

```python
celcius % 2 == 0
```

```python
# matrix manipulation
# matrix product, for element wise use *

A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
print(A * B)

# for matrix product, use @ sign
A @ B
```

```python
# the product of two matrices is only plausible when the inner 
# dimension of the two matrices are the same. 
# The dimensions refer the the n of elements 
# both horizontally and vertically
A.shape
```

```python
# when manipulating the arrays of different types, the type
# of the resulting array will correspond to the more general 
# of the two types
# This is called upcasting 
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[7.1, 8.2, 9.1], [10.4, 11.2, 12.3]])
```

```python
# create arr with 15 elements, with a dimension of 3x5
b = np.arange(1, 16, 1).reshape(3, 5)
b
```
